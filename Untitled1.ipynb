```python
from numpy import *


m = 1
M = 60
R = 1
w = 1
k = m/M
lamda = k/(1 + k)
rm = (1 - lamda) * R
rM = lamda*R



def fx(x, y, rm, rM):
    
    A = -w**2 * R**2
    
    B = (x + rM)*rm
    
    C = (y**2 + (x + rM)**2)**(3/2)
    
    D = (x - rm)*rM
    
    E = (y**2 + (x - rm)**2)**(3/2)
    
    return A * (B/C + D/E)

def fy(x, y, rm, rM):
    
    A = -w**2 * R**2
    
    B = y*rm
    
    C = (y**2 + (x + rM)**2)**(3/2)
    
    D = y*rM
    
    E = (y**2 + (x - rm)**2)**(3/2)
    
    return A * (B/C + D/E)


# In[25]:


def f(r, t):
    x = r[0]
    y = r[1]
    u = r[2]
    v = r[3]
    
    xdot = u
    ydot = v
    
    udot = fx(x,y,rm,rM) + 2*w*v + (w**2) * x
    
    vdot = fy(x,y,rm,rM) - 2*w*u + (w**2) * y
    
    return array([xdot, ydot, udot, vdot], float)


# In[26]:


def canon_ball(t0, tf, N, xinit, yinit):
    
    """This function solves the canon ball equations of motion with air resistance using Runge Kunta"""
    
    h = (tf - t0)/N
    
    tpoints = arange(t0, tf, h)
    xpoints = []
    ypoints = []
    upoints = []
    vpoints = []
    
    r = array([xinit, yinit, 0, 0])
    
    for t in tpoints:
        
        #print(t)
        xpoints.append(r[0])
        #print(xpoints)
        ypoints.append(r[1])
        print(ypoints)
        upoints.append(r[2])
        vpoints.append(r[3])
        
        k1 = h * f(r, t)
        
        k2 = h * f(r + 0.5 * k1, t + 0.5*h)
        
        k3 = h * f(r + 0.5 * k2, t + 0.5*h)
        
        k4 = h * f(r + k3, t + h)
        
        new_term = (k1 + 2*k2 + 2*k3 + k4)/6
        
        r += (k1 + 2*k2 + 2*k3 + k4)/6
    
    
    return tpoints, xpoints, ypoints, upoints, vpoints
        


# In[32]:


#x4 = (R/2) * (1 - k)/(1 + k)
#y4 = R*sin(pi/3)

x4 = R/2
y4 = R/2*sqrt(3)

t, x, y, u, v = canon_ball(0, 80, 100, x4, y4)

x = x + ones(len(x))*rM

from matplotlib import pyplot as plt



plt.plot(x,y)
plt.show()
```
